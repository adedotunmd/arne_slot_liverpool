```python
# Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
from PIL import Image
```

```python
# Import Dataset
import kagglehub

# Download latest version
path = kagglehub.dataset_download("evangower/english-premier-league-standings")

print("Path to dataset files:", path)
```

```python
import os

files = os.listdir(path)
csv_files = [f for f in files if f.endswith('.csv')]

if csv_files:
    # Assuming there's at least one CSV file, take the first one.
    data_file_path = os.path.join(path, csv_files[0])
    df = pd.read_csv(data_file_path)
    print(df.head())
else:
    print("No CSV files found in the dataset directory.")
```

```python
# Extract Liverpool's Data
lfc = df.query('team == "Liverpool"')
lfc
```

```python
# Add the last 2 seasons manually

# 1. Create a dictionary with the 2 new rows
new_data = {
    'season_end_year': ['2025', '2026'],
    'team': ['Liverpool', 'Liverpool'],
    'position': [1, 5],
    'played': [38, 38],
    'won': [25, 17],
    'drawn': [9, 9],
    'lost': [4, 12],
    'gf': [86, 63],
    'ga': [41, 53],
    'gd': [45, 10],
    'points': [84, 60],
      'notes': ['→ Champions League via league finish', '→ Champions League via league finish']
}


# 2. Convert the new data to a DataFrame
df_new = pd.DataFrame(new_data)

# 3. Concatenate them vertically
lfc_new = pd.concat([lfc, df_new], ignore_index=True)
```

```python
# Liverpool's Inconsistent Trend over the years

lfc_new['season_end_year'] = pd.to_numeric(lfc_new['season_end_year']) # Ensure season_end_year is numeric

fig = px.scatter(
    lfc_new,
    x='season_end_year',
    y='points',
    title="Liverpool's Inconsistency in the Premier League Era",
    color='position',  # Color the line based on position
    color_continuous_scale='RdYlGn_r', # Green for low position, Red for high position
    hover_data={'position': True} # Show position on hover
)

fig.update_layout(
    xaxis_showgrid=True,
    yaxis_showgrid=True,
    template="plotly_white"
)

fig.update_traces(
    mode="lines+markers" # Show markers at each data point
)
img = Image.open("/content/drive/MyDrive/Sports Analytics/Stats_vs_Eye_Test_removebg.png")

fig.add_layout_image(
    dict(
        source=img,
        xref="paper",
        yref="paper",
        x=1.15,          # move slightly beyond right edge
        y=-0.20,         # move slightly below plot
        xanchor="right",
        yanchor="bottom",
        sizex=0.21,
        sizey=0.21,
        sizing="contain",
        opacity=0.8,
        layer="above"
    )
)

fig.show()
```

```python
#League Position over time

# 1. Create a basic line chart with markers
fig2 = px.line(
    lfc_new,
    x='season_end_year',
    y='position',
    markers=True,
    title="Liverpool's League Position Over Time",
    labels={'position': 'League Finish', 'season_end_year': 'Season'}
)

# 2. Crucial step: Invert the Y-axis so 1st place is at the top
fig2.update_yaxes(autorange="reversed", tickmode='linear', dtick=1)

# 3. Add a shaded background area for the Top 4 (Champions League spots)
fig2.add_hrect(
    y0=0.5, y1=4.5, # Covers positions 1 through 4
    fillcolor="lightgreen", opacity=0.2,
    layer="below", line_width=0,
    annotation_text=" Top 4",
    annotation_position="bottom left"
)

fig2.update_layout(
    template="plotly_white",
    xaxis_showgrid=True,
    yaxis_showgrid=False
)

fig2.add_annotation(
    x=2026,
    y=5,
    text="Ucl Spot",
    showarrow=False,
    xanchor="left",
    yanchor="top",
    font=dict(color="black")
)

img = Image.open("/content/drive/MyDrive/Sports Analytics/Stats_vs_Eye_Test_removebg.png")

fig2.add_layout_image(
    dict(
        source=img,
        xref="paper",
        yref="paper",
        x=1.13,          # move slightly beyond right edge
        y=-0.12,         # move slightly below plot
        xanchor="right",
        yanchor="bottom",
        sizex=0.23,
        sizey=0.23,
        sizing="contain",
        opacity=0.8,
        layer="above"
    )
)

fig2.show()
```

```python
#Import Injuries Dataset

lfc_injuries = pd.read_csv('/content/drive/MyDrive/Sports Analytics/liverpool_injuries.csv', encoding='latin1')
```

```python
lfc_injuries.dropna(inplace=True)
lfc_injuries
```

```python
# Number of Days injured by Season

season_total_injuries = lfc_injuries.groupby('Season')['Days Injured'].sum().reset_index()

fig = px.bar(
    season_total_injuries,
    x='Season',
    y='Days Injured',
    text='Days Injured',
    title="Were Injuries a Factor? Examining Liverpool's Three Dip Seasons"
)

fig.update_layout(
    xaxis_showgrid=True,
    yaxis_showgrid=True
)

fig.update_layout(
    template="plotly_white"
)

fig.update_traces(
    textposition="outside"
)

img = Image.open("/content/drive/MyDrive/Sports Analytics/Stats_vs_Eye_Test_removebg.png")

fig.add_layout_image(
    dict(
        source=img,
        xref="paper",
        yref="paper",
        x=1.13,          # move slightly beyond right edge
        y=-0.12,         # move slightly below plot
        xanchor="right",
        yanchor="bottom",
        sizex=0.23,
        sizey=0.23,
        sizing="contain",
        opacity=0.8,
        layer="above"
    )
)

fig.show()
```

```python
# Number of Injuries by season

injured_players_count = lfc_injuries.groupby('Season')['Name'].nunique().reset_index()
injured_players_count.rename(columns={'Name': 'Injured Players'}, inplace=True)

fig = px.bar(
    injured_players_count,
    x='Season',
    y='Injured Players',
    text='Injured Players',
    title='Number of Unique Injured Players Per Season'
)

fig.update_layout(
    xaxis_showgrid=True,
    yaxis_showgrid=True
)

fig.update_layout(
    title="Liverpool's Three Dip Seasons: Total Number of Injured Players",
    template="plotly_white"
)

fig.update_traces(
    textposition="outside"
)

img = Image.open("/content/drive/MyDrive/Sports Analytics/Stats_vs_Eye_Test_removebg.png")

fig.add_layout_image(
    dict(
        source=img,
        xref="paper",
        yref="paper",
        x=1.13,          # move slightly beyond right edge
        y=-0.12,         # move slightly below plot
        xanchor="right",
        yanchor="bottom",
        sizex=0.23,
        sizey=0.23,
        sizing="contain",
        opacity=0.8,
        layer="above"
    )
)

fig.show()
```
